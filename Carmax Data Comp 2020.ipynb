```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#loading in data
df = pd.read_csv('CaseCompetitionData.csv')

#renaming headers / dropping not needed ones
df.columns = ['insertNum','make','model','vehicleYear','price','tradeIn','financing','age','income','gender','previousPurchase'
,'distanceToDealer','postPurchaseSatisfaction','warranty','subsequentPurchases']

dfEdited = df.drop(columns=['insertNum','gender','distanceToDealer','postPurchaseSatisfaction'])

#Sort into low income levels
dfEdited['income'].replace(['0 - 20000','20001 - 40000'], np.nan,regex=True, inplace=True)
dfEdited['income'].fillna('Low', inplace=True)

#Sort into mid income levels
dfEdited['income'].replace(['40001 - 60000','60001 - 80000','80001 - 100000','100001 - 120000','120001 - 140000','140001 - 160000'], np.nan,regex=True, inplace=True)
dfEdited['income'].fillna('Mid', inplace=True)

#Sort into high income levels
dfEdited['income'].replace(['160001 - 180000','180001 - 200000','200001+'], np.nan,regex=True, inplace=True)
dfEdited['income'].fillna('High', inplace=True)
dfEdited



```
